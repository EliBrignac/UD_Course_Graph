```python
#Creating the Course object
class Course:
    def __init__(self, name, title, credits, description, prereqs, after_courses, restrictions, typically_offered):
        self.name = name
        self.title = title
        self.credits = credits
        self.prereqs = prereqs  # dictionary of a list of course objects and a string
        self.after_courses = after_courses
        self.restrictions = restrictions  # String probly or list of course objects
        self.description = description
        self.typically_offered = typically_offered

    def get_name(self):
        return self.name

    def get_title(self):
        return self.title

    def get_credits(self):
        return self.credits
    
    def get_description(self):
        return self.description

    def get_prereqs(self):
        return self.prereqs
    
    def after_courses(self):
        return self.after_courses

    def get_restrictions(self):
        return self.restrictions

    def get_typically_offered(self):
        return self.typically_offered
    
# Alot of attributes of this course object aren't used in this code such as the restrictions, credits, typically offered, etc.
# It just wasn't alot of work to parse them all so I just thought I would parse them all anyways. Also some courses don't even
# have these attributes like CISC 473 doesn't have a typically_offered or prereq listed on the UDel course Catalog. 


    
    
# link to find lots of courses on the same page
# https://catalog.udel.edu/content.php?catoid=63&catoid=63&navoid=16384&filter%5Bitem_type%5D=3&filter%5Bonly_active%5D=1&filter%5B3%5D=1&filter%5Bcpage%5D=37#acalog_template_course_filter
start_time = time.time()


# Global Dictionary that will be filled with all the course objects
dictionary_of_all_courses = {}

```

```python

def get_text_on_page(url):
    '''
    Geting the text on the page with beautiful soup.
    
    Parameters:
        url (raw str): the url you want to find the courses on. It MUST be the url 
        listed below with one of the course types chosen.
        
        Course type is chosen either on page, or by this key ↓ in the url, in this example its ACCT.
                                                             ↓
                                                             ↓
        https://catalog.udel.edu/content.php?filter%5B27%5D=ACCT&filter%5B29%5D=&filter%5Bcourse_type%5D=-1&filter%5Bkeyword%5D=&filter%5B32%5D=1&filter%5Bcpage%5D=1&cur_cat_oid=63&expand=&navoid=16384&search_database=Filter#acalog_template_course_filter
        
        If its not a url like this then the whole thing won't work
        
    returns:
        text (str) : the info gotten from the html
    '''

    html = urllib.request.urlopen(url)
    # parsing the html file
    htmlParse = BeautifulSoup(html, 'html.parser')

    text = str(htmlParse.get_text)
    
    # prints the text it gets for debugging purposes
    #print(url)
    #print(text)
    return text
```

```python

def parse_prereqs(info):
    '''
    does NOT yet work for all course types. For instance, MATH 210 is a prereq for CISC 304, but this will not be shown 
    in any graph because they are two different course types. How I would fix this is include a list of all possible 
    course types as a varaibel, and then just check for those course types in the pattern detection part of this code (this part
    re.findall(course_type + ' ' + r'\b\d{3}\b', info)). Same goes for the parse_restrictions() function. I can hook this up if 
    needed but it makes the graphs very large especially for lower courses (for example MATH 241)
     
    '''
    
    # Returns a dictionary with a list of the prereqs as nodes, and the information for the prereqs
    course_type = info[0: 4] #CISC or MATH or ANTH or whatever
    #print(course_type)
    # Gets the name of the course so it doesn't get added to the prereqs
    course_name = info[0: info.find(' - ')]
    # parse the info to only be the prereqs
    if info.find('RESTRICTIONS: ') != -1:
        info = info[info.find('PREREQ: '): info.find('RESTRICTIONS: ')]
    elif info.find('Course Typically Offered:') != -1:
        info = info[info.find('PREREQ:'): info.find('Course Typically Offered:')]

    print('BEFORE:', info)
    print('NORMALIZED:', info.replace('/xa0', ' '))
    prereq_list = []

    # Find the names of the courses within the prereq
    prereq_names = []
    for prereq in re.findall(course_type + ' ' + r'\b\d{3}\b', info):
        if (prereq not in prereq_names) and (prereq != course_name):
            print('PARSE: ', prereq, course_name)
            prereq_names.append(prereq)

    # print(info)
    # print(prereq_list)
    # Turn all of the prereqs into Nodes
    for prereq in prereq_names:
        prereq = pydot.Node(prereq)
        prereq_list.append(prereq)

    # What the prereq actually says
    description = info

    return {'Courses': prereq_list, 'Description': description}
```

```python

def parse_restrictions(info):
    course_type = info[0: 4]
    info = info[info.find('RESTRICTIONS:'): info.find('Course Typically Offered:')]
    restriction_list = []
    restriction_names = re.findall(course_type + ' ' + r'\b\d{3}\b', info)

    for restriction in restriction_names:
        restriction = pydot.Node(restriction)
        restriction_list.append(restriction)

    description = info
    return {'Courses': restriction_list, 'Description': description}

```

```python

def parse_credits(info):
    info = info[info.find('Credit(s):') + 11:]
    title = info[info.find(' - ') + 3: info.find('Credit(s)') - 1]
    if info.find(title.upper()) == -1:
        for char in info:
            if char.lower() in ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q',
                                'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']:
                position = info.find(char)
                credits = info[:position]
                break
    else:
        credits = info[info.find('Credit(s):') + 11: info.find(title.upper())]

    return credits
```

```python

def parse_course_info(info):
    print('INFO STUFF:',info)
    # Get the course name and make it a node
    name = pydot.Node(info[0: info.find(' - ')])
    print('Name:', name)

    # Get the course Title
    title = info[info.find(' - ') + 3: info.find('Credit(s)') - 1]
    print('Title:', title)
    # Get the credits the course is worth
    credits = parse_credits(info)
    print('Credits:', credits)

    # Get the description
    description = info[info.find('Component:'): info.find('Repeatable for Credit:')]
    print('Description:', description)

    # Get the prereqs, its a dictionary, look at the parse_prereqs function documentation
    prereqs = parse_prereqs(info)
    print('Prereqs:', prereqs)

    # Gets the after courses to an empty list that will eventually be filled with Nodes
    after_courses = []
    print('After Courses:', after_courses)

    # Gets the restrictions of the course. A dictionary with the list of restricted courses
    # and then the description of the restrictions
    restrictions = parse_restrictions(info)
    print('Restrictions:', restrictions)

    # Gets when the course is typically offered:
    typically_offered = info[info.find('Course Typically Offered:'): info.find('General Education Objectives:')]
    print('Typically Offered:', typically_offered)
    print('\n')

    return (name, title, credits, description, prereqs, after_courses, restrictions, typically_offered)

```

```python

def find_after_courses(course_name):
    # finds the after courses for all of the course name passed to it
    course_obj = dictionary_of_all_courses[course_name]

    for course in dictionary_of_all_courses:
        for prereq in dictionary_of_all_courses[course].get_prereqs()['Courses']:
            prereq_name = prereq.get_name()[1:-1]
            if course_name == prereq_name:
                course_obj.after_courses.append(dictionary_of_all_courses[course].get_name())
    print(course_name)
    print(course_obj.after_courses)
    return
```

```python

def build_tree(G, course_name):
    '''
    This is a function that will build the after_course graph of a particular course. 
    It is a recurssive function that will loop through until it is finished.
    
    Parameters:
        G (pydot.Dot object): G is an empty pydot digraph that is strict for this function to work
            as intended: G = pydot.Dot(graph_type = "digraph", strict = True)
            
            I suppose it doesn't have to be strict or a digraph but the end 
            result looks better if it is. 
    
        course_name (str): the name of the course you want to build the graph of as
                        a string. Example: 'MATH 241'
    
    returns:
        G (pydot.Dot object): the completed graph of the after_courses
    
    '''
    #GRAPH
    
    course_node = dictionary_of_all_courses[course_name].get_name()
    
    for node in G.get_nodes():
        if course_name == node.get_name()[1:-1]:
            course_node = node
    
    G.add_node(course_node)
    
    for after_course in dictionary_of_all_courses[course_name].after_courses:
        
        node1 = pydot.Node(after_course.get_name()[1:-1])
        
        G.add_node(node1)
        
        edge1 = pydot.Edge(course_node, node1)
        
        G.add_edge(edge1)
        
        build_tree(G, after_course.get_name()[1:-1])
        
        
    return G
```

```python

def get_courses_listed_on_page(text, course_type):
    
    # A loop for all the courses listed on the webpage
    # Gets all the courses
    print(text.count(';coid'))
    for i in range(0, text.count(';coid')):
        # Gets the number of the course via the html
        html_num = text[text.find("'63',") + 7: text.find(',this,') - 1]
        text = text[text.find(',this,') + 7:]
        print(html_num)

        # Search the specific course URL
        course_url = 'https://catalog.udel.edu/preview_course_nopop.php?catoid=63&coid=' + str(html_num)
        course_html = urllib.request.urlopen(course_url)
        course_htmlParse = BeautifulSoup(course_html, 'html.parser')

        # Parses the website for all of the courses and their information
        for para in course_htmlParse.find_all('td', {'class': 'block_content'}):
            info = str(para.get_text())
            # prints the parsed information of the courses
            #Course type examples: CISC, MATH, STAT, ANTH, etc.
            info = info[info.find(course_type): info.find('Back to Top | ')].strip('\n').strip("'").strip('"')
            # print(info[info.find('MATH'): info.find('Back to Top | ')].strip('\n').strip("'").strip('"'))

            # Create Course objects from the parsed course info
            name, title, credits, description, prereqs, after_courses, restrictions, typically_offered = parse_course_info(info)

            course = Course(name, title, credits, description, prereqs, after_courses, restrictions, typically_offered)
            # Adds the courses to the course dictionary
            dictionary_of_all_courses[course.get_name().get_name()[1:-1]] = course

    return
```

```python

def format_prereq_description(description):
    new_str = ''
    helper = 0
    parse_length = 40
    if len(description) > parse_length:
        for i in range(1, len(description)):
            if i % 40 == 0:
                if description[i] == ' ':
                    new_str += description[helper:i] + '\n'
                    helper = i
                else:
                    for j in range(i, len(description)):
                        if description[j] == ' ':
                            new_str += description[helper:j] + '\n'
                            helper = j + 1
                            break
    else:
        new_str = description
    return new_str

```
